```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("country_wise_latest.csv")

top_confirmed = df.sort_values(by="Confirmed", ascending=False).head(10)
top_deaths = df.sort_values(by="Deaths", ascending=False).head(10)
top_recovered = df.sort_values(by="Recovered", ascending=False).head(10)


```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Confirmed", y="Country/Region", data=top_confirmed, palette="Oranges_r")
plt.title("Top 10 Countries by Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
#  Plot 2: Top 10 Countries by Deaths
plt.figure(figsize=(10, 6))
sns.barplot(x="Deaths", y="Country/Region", data=top_deaths, palette="Reds_r")
plt.title("Top 10 Countries by Deaths")
plt.xlabel("Death Count")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Recovered", y="Country/Region", data=top_recovered, palette="Greens_r")
plt.title("Top 10 Countries by Recovered Cases")
plt.xlabel("Recovered Cases")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
#  Plot 4: Global COVID-19 Distribution Pie Chart
global_sums = df[['Recovered', 'Active', 'Deaths']].sum()
plt.figure(figsize=(6, 6))
plt.pie(global_sums, labels=global_sums.index, autopct='%1.1f%%',
        colors=['green', 'orange', 'red'], startangle=140)
plt.title("Global COVID-19 Case Distribution")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#  Plot 5: Heatmap of COVID Stats for Top 20 Countries by Confirmed
df_heat = df[['Country/Region', 'Confirmed', 'Deaths', 'Recovered', 'Active']].set_index('Country/Region')
df_heat_top20 = df_heat.sort_values(by='Confirmed', ascending=False).head(20)
plt.figure(figsize=(12, 8))
sns.heatmap(df_heat_top20, annot=True, fmt='.0f', cmap="YlOrRd", linewidths=.5)
plt.title("Heatmap: Top 20 Countries by COVID-19 Statistics")
plt.xlabel("Metrics")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
